# YOLO 번호판 탐지 모델 학습

본 노트북은 번호판 탐지 모델의 학습 및 추가 학습(Fine-Tuning)에 사용된 코드입니다.

- 1차 학습 : Roboflow 번호판 데이터셋
- 추가 학습 : AI Hub CCTV 번호판 데이터셋
- Framework : Ultralytics YOLO
- Language : Python
- Platform : Google Colab / Windows

```python
# ==========================================
# YOLO 번호판 학습 (Roboflow Dataset)
# ==========================================

!nvidia-smi
!pip install -U ultralytics

from ultralytics import YOLO

# 사전학습 모델 불러오기
model = YOLO("yolo26n.pt")

# 모델 학습
model.train(
    data="data.yaml",
    epochs=50,
    imgsz=640,
    batch=8,
    patience=20,
    name="license_plate_yolo26"
)
```

```python
# ==========================================
# AI Hub 데이터 추가 학습 (Fine-Tuning)
# ==========================================

from ultralytics import YOLO

# 1차 학습된 모델(best.pt) 불러오기
model = YOLO("best.pt")

# AI Hub 데이터 추가 학습
model.train(
    data="data.yaml",
    epochs=5,
    imgsz=960,
    batch=8,
    device=0,
    workers=0,
    name="license_aihub_finetune"
)
```
